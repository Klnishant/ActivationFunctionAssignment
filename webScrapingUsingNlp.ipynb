```python
import requests
import re
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python

import nltk
nltk.download()
```

```python
url = "https://en.wikipedia.org/wiki/Independence_Day_(India)"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
paragraphs = soup.find_all('p')
paragraph_text = [paragraphs.get_text() for paragraphs in paragraphs]
```

```python
paragraph_text
```

```python
cleaned_text = [re.sub(r'[^a-zA-Z0-9\s]','',texts) for texts in paragraph_text]
```

```python
cleaned_text
```

```python
tokenized_text = [word_tokenize(texts) for texts in cleaned_text]
```

```python
stopwords = set(stopwords.words('english'))
stopwords
```

```python
filtered_text = [[word for word in token if word not in stopwords] for token in tokenized_text]
filtered_text
```

```python
stemmer = PorterStemmer()
stemmed_text = [[stemmer.stem(word) for word in token] for token in filtered_text]
stemmed_text
```

```python
final_text = [[word for word in token if word.strip()] for token in stemmed_text]
final_text
```

```python
sentences = [' '.join(tokens) for tokens in final_text]
sentences
```

```python
proccessed_paragraphs = '\n\n'.join(sentences)
proccessed_paragraphs
```
